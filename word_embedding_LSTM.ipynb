```python
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
data['review'].str.len()
```

```python
data['review'] = data['review'].str.lower()
data['review']
```

```python
# removing HTML tags
import re
def remove_html_tags(text):
    re_html = re.compile('<.*?>')
    return re_html.sub(r'', text)
```

```python
data['review'] = data['review'].apply(remove_html_tags)
```

```python
# removing URLS
def remove_url(text):
    re_url = re.compile('https?://\S+|www\.\S+')
    return re_url.sub('', text)
```

```python
data['review'] = data['review'].apply(remove_url)
```

```python
# removing stop words
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_english = stopwords.words('english')


def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords_english:
            continue
        else:
            new_text.append(word)

    return ' '.join(new_text)
```

```python
data['review'] = data['review'].apply(remove_stopwords)
data['review']
```

```python
# stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def perform_stemming(text):
    new_text = [ps.stem(word) for word in text.split()]
    return ' '.join(new_text)
```

```python
data['updated_review'] = data['review'].apply(perform_stemming)
```

```python
len(data['updated_review'].unique())
```

```python
### Vocabulary size
voc_size=50000
```

```python
words = []
```

```python
words = data['updated_review'].copy()
```

```python
words
```

```python
len(words[1])
```

```python
onehot_repr=[one_hot(sent,voc_size)for sent in words] 
print(onehot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## pre padding
sent_length=1000
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## 10 feature dimesnions
dim=10
```

```python
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
# model.compile('adam','mse')
print(model.summary())
```
